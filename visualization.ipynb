```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = sns.load_dataset('iris')
df
```

```python
# barplot for sepal length
_ = sns.barplot(x = 'species', y = 'sepal_length', data = df) # underscore here denotes garbage variable
```

```python
_ = sns.barplot(x = 'species', y = 'petal_length', data = df, estimator = sum) # underscore here denotes garbage variable
```

```python
_ = sns.barplot(x = 'species', y = 'petal_width', data = df, estimator = sum)
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize = (15,10), nrows = 2, ncols = 2)   # Here fig denotes the figure size
_ = sns.barplot(x = 'species', y = 'sepal_length', data = df , ax = ax[0,0])
_ = sns.barplot(x = 'species', y = 'sepal_width', data = df, ax = ax[0,1])
_ = sns.barplot(x = 'species', y = 'petal_length', data = df, ax = ax[1,0])
_ = sns.barplot(x = 'species', y = 'petal_width', data = df, ax = ax[1,1])
```

```python
fig,ax = plt.subplots(figsize = (30,10), ncols = 4)
_ = sns.barplot(x = 'species', y = 'sepal_length', data = df , ax = ax[0])
_ = sns.barplot(x = 'species', y = 'sepal_width', data = df, ax = ax[1])
_ = sns.barplot(x = 'species', y = 'petal_length', data = df, ax = ax[2])
_ = sns.barplot(x = 'species', y = 'petal_width', data = df, ax = ax[3])
```
